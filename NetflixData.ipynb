# Netflix Data

```python
import pandas as pd
import numpy as np
```

# Data Loading

```python
mydata = pd.read_csv('netflix_titles.csv')
```

# Data Cleaning

```python
mydata.head()
```

```python
mydata.isna().sum()
```

```python
mydata['country'] = mydata['country'].fillna(mydata['country'].mode()[0])
```

```python
mydata.head()
```

```python
mydata['cast'].replace(np.nan, 'No Data',inplace  = True)
mydata['director'].replace(np.nan, 'No Data',inplace  = True)
```

```python
mydata.dropna(inplace=True)
mydata.drop_duplicates(inplace=True)
```

```python
mydata.head()
```

```python
mydata.isna().sum()
```

```python
mydata.dtypes
```

```python
mydata["date_added"] = pd.to_datetime(mydata['date_added'])
mydata.head()
```

```python
mydata.dtypes
```

```python
mydata['rating'].unique()
```

```python
ratings = {'TV-PG': 'Older Kids (7+)', 'TV-MA': 'Adults (18+)', 'TV-Y7-FV': 'Older Kids (7+)', 'TV-Y7': 'Older Kids (7+)', 'TV-14': 'Young Adults (16+)', 'R': 'Adults (18+)', 'TV-Y': 'Kids (All)', 'NR': 'Adults (18+)', 'PG-13': 'Teens (13+)', 'TV-G': 'Kids (All)', 'PG': 'Older Kids (7+)', 'G': 'Kids (All)', 'UR': 'Adults (18+)', 'NC-17': 'Adults (18+)'
}

mydata['rating'] = mydata['rating'].replace(ratings)
```

```python
mydata['rating'].unique()
```

```python
mydata.head()
```

```python
mydata['main_country'] = mydata['country'].apply(lambda x: x.split(',')[0])
```

```python
mydata['main_country'].head()
```

```python
mydata.head()
```

```python
mydata['main_country'].unique()
```

```python
mydata.head()
```

# Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = 8,4
```

```python
Country_Counts = mydata.groupby(['main_country','type']).size().unstack().fillna(0)
Country_Counts
```

```python
high_countries = Country_Counts.sum(axis=1).sort_values(ascending=False)[:10]
high_countries
```

```python
high_countries_counts = Country_Counts.loc[high_countries.index]
high_countries_counts
```

```python
high_countries_counts.plot(kind='bar', figsize=(10,6), color = ['red','black'])
plt.title("Movies and TV Shows by top 10 Countries")
plt.xlabel('Country')
plt.ylabel('Count')
plt.legend(['Movie','TV Show'], )
plt.show()
```

```python
rating_country = mydata.groupby(['rating','main_country'])['show_id'].count().reset_index(name='count')
```

```python
top_country = rating_country.groupby('main_country').sum().sort_values(by='count', ascending=False).head(10).index
rating_country = rating_country[rating_country['main_country'].isin(top_country)]
```

```python
pivot = rating_country.pivot(index = 'rating', columns = 'main_country', values = 'count')
pivot
```

```python
sns.heatmap(pivot, cmap='Reds', annot=True, fmt='g', cbar_kws={'label': 'Count'})
```

```python
mydata['year'] = mydata['date_added'].dt.year
mydata.head()
```

```python
year_wise = mydata.groupby(['year']).size()
year_wise
```

```python
year_wise.plot(kind='line', figsize=(10,6), color = 'red')
plt.title("Shows Releases Year wise")
plt.xlabel('Year')
plt.ylabel('No of Shows')
plt.show()
```

```python
type_sep = mydata.groupby(['type']).size()
type_sep
```

```python
plt.bar(type_sep.index, type_sep.values,color = ['red','black'])
plt.title("No of Movies vs No of Shows")
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
_, _, pies = plt.pie(type_sep, labels=['Movie', 'TV Show'], colors=['Red', 'Black'], autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('No of Movies vs No of Shows')
for pie in pies:
    pie.set_color('white')
plt.show()
```
